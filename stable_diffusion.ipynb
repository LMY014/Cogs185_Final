```python
# stable_diffusion_gen.py

import torch
from diffusers import StableDiffusionPipeline
from PIL import Image
import json, os

# ✅ Load captions
with open("/content/cifar10_captions.json") as f:
    captions = json.load(f)

# ✅ Set device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# ✅ Load Stable Diffusion model (you can change this to SDXL or other if needed)
pipe = StableDiffusionPipeline.from_pretrained(
    "CompVis/stable-diffusion-v1-4",
    torch_dtype=torch.float16 if device.type == "cuda" else torch.float32
).to(device)

# ✅ Create output folder
output_dir = "outputs/sd"
os.makedirs(output_dir, exist_ok=True)

# ✅ Generate images
for name, data in list(captions.items())[:10]:  # change limit if needed
    prompt = data["caption"]
    image = pipe(prompt).images[0]
    image.save(os.path.join(output_dir, name))
    print(f"Generated image for: {prompt}")

print(f"\n✅ Done! Images saved in {output_dir}")
```
